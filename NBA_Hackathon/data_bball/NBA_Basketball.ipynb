```python
# SQL for creating Game_Lineup_Data data frame:

query = '''
SELECT *
FROM
    Game_Lineup_Data_Sample
WHERE
    status = 'A'
'''

df_game_lineup = pd.read_sql(query, con=engine_nba_hackathon_basketball)
df_game_lineup
```

```python
# Function for defining starting lineups for each period:

def set_starting_lineup(game_period):
    
    # Initialize the lineup list we'll use for + / -
    lineup = []
    
    # Set the starters from the lineup given in Game_Lineup_Data
    starters = df_game_lineup[(df_game_lineup.Game_id == game_period.Game_id) & (df_game_lineup.Period == game_period.Period)]
    
    # Add the players to our new + / - list with initial values for + / - and active (in-game or not)
    for i, player in starters.iterrows():
        lineup.append({
            'Team_id':player.Team_id,
            'Player_id':player.Person_id,
            'Player_Plus/Minus':0,
            'Active':1
        })
    return lineup
```

```python
# Function for performing a substitution:

def substitution(event,lineup):
    already_played = False
    
    # Substitute players, for existing players:
    for player in lineup:
        if player['Player_id'] == event['Person1']:
            player['Active'] = 0
        elif player['Player_id'] == event['Person2']:
            already_played = True
            player['Active'] = 1

    # If substituted player hasn't played yet, put them in the lineup:
    if already_played == False:
        lineup.append({
            'Team_id':event['Team_id'],
            'Player_id':event['Person2'],
            'Player_Plus/Minus':0,
            'Active':1
        })
        
    return lineup
```

```python
# Function for scoring a field goal:

def score_field_goal(event,lineup):
    
    # Which team scored?
    scoring_team_id = event['Team_id']
    
    for player in lineup:
        
        # for active players 
        if player['Active'] == 1:
            
            # + for scoring team
            if player['Team_id'] == scoring_team_id:
                player['Player_Plus/Minus'] += event['Option1']
    
            # - for opposing team
            else:
                player['Player_Plus/Minus'] -= event['Option1']
                
    return lineup
```

```python
# Function for scoring a free throw:

def score_free_throw(event,lineup):
    
    # Which team scored?
    scoring_team_id = event['Team_id']
    
    for player in lineup:
        
        # Only score when Option1 = 1
        if event['Option1'] == 1:
            
            # for active players
            if player['Active'] == 1:

                # + for scoring team
                if player['Team_id'] == scoring_team_id:
                    player['Player_Plus/Minus'] += event['Option1']

                # - for opposing team
                else:
                    player['Player_Plus/Minus'] -= event['Option1']
```

```python
# Get every game + period combo that we'll need to process:
game_periods = df_play_by_play_clean_sorted[['Game_id','Period']].drop_duplicates()
game_periods
```

```python
df_plus_minus = pd.DataFrame()

# Iterate through all events
for i, game_period in game_periods.iterrows():
    
    # Initialize starting lineup for game period
    lineup = set_starting_lineup(game_period)
    
    # Get all events for that game period
    valid_events = df_play_by_play_clean_sorted[(df_play_by_play_clean_sorted.Game_id == game_period.Game_id) & (df_play_by_play_clean_sorted.Period == game_period.Period)]
    
    # Process every event in the game period
    for i, event in valid_events.iterrows():
        if event.Event_Msg_Type == 1:
            score_field_goal(event,lineup)
        elif event.Event_Msg_Type == 3:
            score_free_throw(event,lineup)
        elif event.Event_Msg_Type ==8:
            substitution(event,lineup)
    
    # Create a dataframe of + / - for the period and add it to a cumulative set
    df_event = pd.DataFrame(lineup)
    df_event.drop(columns='Active',inplace=True)
    df_event['Game_id'] = game_period.Game_id
    df_event['Period'] = game_period.Period
    df_plus_minus = pd.concat([df_plus_minus,df_event])
```
